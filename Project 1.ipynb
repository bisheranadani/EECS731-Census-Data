The purpose of this project is to perform exploratory data analysis and practicing data preparation processes by combining two data sets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
demographics_df = pd.read_csv('./data/Demographic_Statistics_By_Zip_Code.csv')
```

```python
demographics_df.head()
```

```python
wages_df = pd.read_csv('./data/free-zipcode-database.csv', low_memory=False)
```

```python
wages_df.head()
```

## Two datasets were imported here. One that contains demographic data by zipcode, and one the contains wages by zipcode.

To attempt to clean up the wages dataset, I will remove the columns 2,3,5,11,19 and all records with NaN for TotalWages. 

```python
wages_df.drop(wages_df.columns[[2,3,5,11,19]], axis=1, inplace=True)
wages_df.dropna()
```

## Visualizing the mean income by state

```python
wages_df.groupby('State')['TotalWages'].mean().plot(kind='bar', figsize=(20,10))
```

### Total wages by state

## Since the demographic dataset contains information only for the State of New York, I extracted data from the wages dataset for NY only

```python
wages_df_NY = wages_df.loc[wages_df['State']=='NY']
wages_df_NY.shape
```

## Combining the two data sets can yield very interesting results in terms of how demographics affect wages in different zipcodes within different states. Combined race, sex, and, number of people on public assistance data with total wages for each zipcode provides insight on the needs of specific communities.

```python
combined_df = demographics_df.merge(wages_df_NY, left_on='JURISDICTION NAME', right_on='Zipcode', how='outer')
```

```python
combined_df.head()
```

```python
combined_df.shape
```

## The merging of these two data sets is not sufficient in its current form and needs more clean-up. There is a discrepancy between the two sets were demographics_df contains data for each zipcode, of which there are only 267 because the data set from the census bureau is incomplete, and wages_df contains data for different cities within the same zipcode. Ideally, complete zipcode data would be obtained and merged in such a way that the zipcode doesn't duplicate for each city (Multi-Index table). 

```python
combined_df.groupby('JURISDICTION NAME')['TotalWages', 'PERCENT MALE'].mean().plot(x='PERCENT MALE', y='TotalWages',kind='scatter')
```

## The wages dataset also contains multiple strange duplicates where the coordinates and wage estimates are the same for multiple cities. This largely skews results and creates a scatter plot like the one above where Percent Male is calculated as 0 but it still shows high results for total wages.

```python
combined_df_v2 = combined_df.loc[combined_df['PERCENT MALE']!=0]
```

```python
combined_df_v2.shape
```

```python
combined_df_v2.groupby('Zipcode')['TotalWages', 'PERCENT MALE'].mean().plot(x='PERCENT MALE', y='TotalWages',kind='scatter')
```

### This graph depicts how the percentage of males in a zipcode affects total wages

## Alternatively, we can try to look at data for fewer zipcodes

```python
most_common_zip = combined_df_v2['Zipcode'].value_counts().idxmax()
print(most_common_zip)
```

```python
combined_df_v2.loc[combined_df_v2['TotalWages'] > 750000000].groupby('Zipcode')['PERCENT US CITIZEN','PERCENT RECEIVES PUBLIC ASSISTANCE'].mean().plot(x='PERCENT US CITIZEN',y='PERCENT RECEIVES PUBLIC ASSISTANCE',kind='scatter')
```

### For cities that have total wages greater than 750000000, how percentage of US citizens correlates with percentage of people that receive public assistance
